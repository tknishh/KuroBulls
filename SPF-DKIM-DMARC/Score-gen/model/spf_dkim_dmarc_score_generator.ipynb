```python
import pandas as pd
```

```python
headers = ['spf','dkim','dmarc']
df = pd.read_csv('/content/spf-dkim-dmarc-vals.csv', header = None, names = headers)
```

```python
df.head(3)
```

```python
def spf_score_gen(data):
  lst = []
  for i in range(len(data)):
    a = data['spf'][i]*data['dkim'][i]*data['dmarc'][i]
    lst.append(a)
  return lst
```

```python
data = df
data['score'] = spf_score_gen(data)
```
